```python
import pandas as pd
from pathlib import Path

filePath = Path('lyrics', 'lyrics_tokenized_v1.csv')

df = pd.read_csv(filePath, sep =';')
df.drop('id', axis = 'columns', inplace = True)
df.head(7)
```

```python
# get all words with two letters
two = df[df["lyrics"].str.len() < 3]
two['lyrics'].unique()
```

```python
# view the most common ones
two = two.groupby(['lyrics']).size().reset_index(name='count').sort_values(by=['count'], ascending=False)
two.head(30)
```

```python
# get all words with two letters
three = df[df["lyrics"].str.len() == 3]
three['lyrics'].unique()
```

```python
# words to be removed from the data
forbidden_words = ['ll', 'nt', 'em', 'co', 'ym', 'ta', 've',
       'se', 'mm', 'uh', 'wo', 'ap', 'fn', 'ba', 'la', 'wi', 
       'fi', 'td', 'cb', 'gs', 'al', 'fe', 'kb', 'jt', 'es', 
       'lo', 'kd', 'um', 'bo', 'du', 'ac', 'da', 'ol', 'li',
       'ii', 'di', 'aw', 'ls', 'mi', 'ot', 'ns', 'rr', 'zs', 
       'xl', 'tj', 'de', 'fo', 'ga', 'st', 'mo', 'qs', 'ps', 
       'jo', 'hе', 'gt', 'el', 'sr', 'pa', 'od', 'hm', 'rb', 
       'ei', 'ny', 'im', 'su', 'tó', 'az', 'sk', 'the']
```

```python
# remove the chosen words
df_cleaned = df[~df["lyrics"].isin(forbidden_words)]

# check if the chosen words have been removed 
df_cleaned[df_cleaned["lyrics"].str.len() < 3]["lyrics"].unique()
```

```python
df_cleaned.to_csv('lyrics_tokenized_v2.csv', sep=";",index=True,index_label="id")
```
